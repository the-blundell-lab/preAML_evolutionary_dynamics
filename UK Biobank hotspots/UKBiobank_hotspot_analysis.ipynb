# Code for UK Biobank figure in UKCTOCS paper

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# import data: UKB participants with at least one CH hotspot
ukb_variants = pd.read_csv('Data_files/CH_double_mutants_AML_anonymised.tsv', sep='\t')
```

```python
# plot
matplotlib.rcParams['pdf.fonttype'] = 42
no_aml = ukb_variants[~ukb_variants['aml_status']]
with_aml = ukb_variants[ukb_variants['aml_status']]

fig, ax = plt.subplots(1, 1, figsize=[5, 5])
ax.scatter(no_aml.VAF, no_aml.VAF2, color=[0.4, 0.4, 0.4], alpha=0.3, edgecolors= "grey", s=70, marker='X')
ax.scatter(with_aml.VAF, with_aml.VAF2, color=(1.0, 0.18, 0.33), edgecolors= "w", s=150, marker='X')

#'#ff0266'

ax.set_xlim(-0.01, 0.52)
ax.set_ylim(-0.02, 0.52)
#ax.plot([0, 1], [0, 1], ls='dashed', color='k', alpha=0.2)
ax.vlines(0.3, -0.5, 1, color='k', alpha=1, lw=1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xticks(np.arange(0, 0.6, 0.1), labels=np.arange(0, 60, 10), fontsize=20)
ax.set_yticks(np.arange(0, 0.6, 0.1), labels=np.arange(0, 60, 10), fontsize=20)
ax.set_xlabel('largest VAF (%)', size=20)
ax.set_ylabel('second largest VAF (%)', size=20)
print(len(with_aml))
plt.tight_layout()
plt.show()
```
